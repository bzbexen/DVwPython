## 2: Plot 12 most dangerous cities

```python
kill.columns
```

```python
city_list = (kill['city'].value_counts().head(12))
city_list
```

```python
city_list.index
```

```python
city_list.values
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x=city_list.index, y=city_list.values ).set_title('Most Dangerous Cities',
                                                              fontdict= { 'fontsize': 18,'fontweight':'bold','color':'blue'})
plt.xticks(rotation=45)
for p in ax.patches: 
    ax.annotate(p.get_height(), (p.get_x()+0.3 , p.get_height()+0.3 ) );
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(10, 8))

sns.countplot(x='city', data=kill, order=kill.city.value_counts().iloc[:12].index).set_title("Most Dangerous Cities",
                                                                                             fontdict= {'fontsize': 18,'fontweight':'bold','color':'blue'})
plt.xticks(rotation=45)

for p in ax.patches: 
    ax.annotate(p.get_height(), (p.get_x()+0.3 , p.get_height()+0.3 ) );
```

## 3: Plot 20 most dangerous states

```python
kill.columns
```

```python
state_list =kill.state.value_counts()[:20]
state_list
```

```python
# with seaborn barplot

fig, ax = plt.subplots(figsize=(10, 8))

sns.barplot(x=state_list.index , y=state_list.values).set_title("Most Dangerous States",fontdict= {'fontsize': 18,'fontweight':'bold','color':'blue'})
plt.xticks(rotation=45)
for p in ax.patches: 
    ax.annotate(p.get_height(), (p.get_x()+0.3 , p.get_height()+0.3 ) );
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(10, 8))

sns.countplot(x="state", data=kill, order= kill.state.value_counts().iloc[:20].index ).set_title("Most Dangerous States",fontdict= {'fontsize': 18,'fontweight':'bold','color':'blue'})
plt.xticks(rotation=45)
for p in ax.patches: 
    ax.annotate(p.get_height(), (p.get_x()+0.3 , p.get_height()+0.3 ) );
```
