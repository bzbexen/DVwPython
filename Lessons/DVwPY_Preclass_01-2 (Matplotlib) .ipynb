## Functional Method

- verilen 4 grafigi ayni yerde gormek istiyoruz 

```python
x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

plt.subplot(2,2,1) # 1. grafik 
plt.plot(x,y1, 'ro'); # x ekseni, y1 sutunu, renk kirmizi, kesisim o ile goster 
plt.title('$y_1=x$');

plt.subplot(2,2,2) # 2. grafik 
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$');

plt.subplot(2,2,3) # 3. grafik 
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(2,2,4) # 4. grafik 
plt.plot(x,y4, 'ks');
plt.title('$y_4=\sqrt{x}$');
```

```python
x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

# plt.figure(figsize=(14,10))
plt.subplot(4,1,1)  # 4'e 1'lik 1. grafik 
plt.plot(x,y1, 'ro');
plt.title('$y_1=x$'); # alt cizgi olunca y'nin altina yazar 

plt.subplot(4,1,2)
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$'); # ^ isareti olursa grafikte us olarak yazar 

plt.subplot(4,1,3)
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(4,1,4)
plt.plot(x,y4, 'ks');
plt.title('$y_4=\sqrt{x}$');
```
